```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(64).reshape(8, 8),columns=["A_A_A","A_A_B","A_B_A","A_B_B","B_A_A","B_A_B","B_B_A","B_B_B"],index=["1_1_1","1_1_2","1_2_1","1_2_2","2_1_1","2_1_2","2_2_1","2_2_2"])
#df = pd.DataFrame(np.arange(64).reshape(8, 8),columns=["A_A_A","A_A_B","A_B_A","A_B_B","B_A_A","B_A_B","B_B_A","B_B_B"])


print(df)
```

```python

def makeMultiColumns(colmun):
    level1=[]
    level2=[]
    level3=[]
    for col in colmun:
        cols=col.split("_")
        level1.append(cols[0])
        level2.append(cols[1])
        level3.append(cols[2])
    return [level1,level2,level3]
col=makeMultiColumns(df.columns)
ind=makeMultiColumns(df.index)
print(ind)
df.columns=col
df.index=ind
```
